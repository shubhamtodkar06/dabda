MERGE DATA

```python
# Define a dictionary containing employee data
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K1', 'K0', 'K1'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],}

# Define a dictionary containing employee data
data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K0', 'K0', 'K0'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']}

# Convert the dictionary into DataFrame
df = pd.DataFrame(data1)

# Convert the dictionary into DataFrame
df1 = pd.DataFrame(data2)


print(df, "\n\n", df1)
```

```python
# using keys from right frame
res = pd.merge(df, df1, how="right", on=['key', 'key1'])

res
```

```python
# using keys from left frame
res = pd.merge(df, df1, how='left', on=['key', 'key1'])

res
```

```python
res = pd.merge(df, df1, how='inner', on=['key', 'key1'])

res
```

SHAPE / RESHAPE DATAFRAMES

```python
#The stack() method works with the MultiIndex objects in DataFrame, it returns a DataFrame with an index with a new inner-most level of row labels. It changes the wide table to a long table.
df_stacked = df.stack()
```

```python
df_stacked
```

```python
# The unstack() is similar to stack method, It also works with multi-index objects in dataframe, producing a reshaped DataFrame with a new inner-most level of column labels.
df_unstacked = df.unstack()
```

```python
df_unstacked
```

```python
# The melt() in Pandas reshape dataframe from wide format to long format. It uses the id_vars[col_names] to melt the dataframe by column names.
df_melt = df.melt()
df_melt
```

```python
plt.scatter(fb_data['Page total likes'], fb_data['like'])
plt.xlabel('Page total likes')
plt.ylabel('Likes')
plt.title('Scatter plot of Page total likes vs. Likes')
plt.show()
```

```python
# Bar chart
sns.countplot(x='Type', data=fb_data)
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Bar chart of Type')
plt.show()
```

```python
# Histogram
plt.hist(fb_data['comment'])
plt.xlabel('Comments')
plt.ylabel('Frequency')
plt.title('Histogram of Comments')
plt.show()
```

```python
# Box plot
sns.boxplot(x='Category', y='Total Interactions', data=fb_data)
plt.xlabel('Category')
plt.ylabel('Total Interactions')
plt.title('Box plot of Total Interactions by Category')
plt.show()
```
